## Part 1 ‐ Exploratory data analysis

The attached logins.json file contains (simulated) timestamps of user logins in a particular
geographic location. Aggregate these login counts based on 15­ minute time intervals, and
visualize and describe the resulting time series of login counts in ways that best characterize the
underlying patterns of the demand. Please report/illustrate important features of the demand,
such as daily cycles. If there are data quality issues, please report them.

```python
import pandas as pd
import numpy as np
import json
import seaborn as sns
from matplotlib import pyplot
from sklearn.tree import DecisionTreeClassifier
import calendar
import matplotlib.cm as cm
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
logins_df = pd.read_json('logins.json', orient='values', convert_dates = True)
```

```python
logins_df.info()
```

There doesn't appear to be missing values in the dataset. 

```python
logins_df.head(3)
```

```python
#Aggregate logins based on 15 minuite time intervals. 
logins_df['count'] = 1
login_agg = logins_df.set_index('login_time').resample('15T').sum()
login_agg = login_agg.reset_index()
login_agg.head(3)
```

```python
print(login_agg.login_time.min())
print(login_agg.login_time.max())
```

There appears to be about four months worth of data. However at a glance it seems that april only goes up until the 13th. 

```python
#Create time features for easier plotting
login_agg['month'] = login_agg.login_time.dt.month
login_agg['week'] = login_agg.login_time.dt.week
login_agg['day'] = login_agg.login_time.dt.day
login_agg['hour'] = login_agg.login_time.dt.hour
login_agg['weekday'] = login_agg.login_time.dt.weekday
```

```python
login_agg.head(3)
```

```python
#Plot full dataset
login_agg.plot(x='login_time', y='count', style='ko-', figsize=(10,5))
pyplot.show()
```

The plot above shows the whole dataset. Its overplotted, but we can still see there seem to be login spikes at a regular interval.

```python
login_agg.groupby('month')['count'].sum().plot(kind='bar')
```

There seems to be a steady increase in the amount of logins per month. The drop in April is probably explained by the 13 days worth of data that was collected. 

```python
ax = login_agg.groupby('weekday')['count'].sum().plot(kind='bar').set_xticklabels([calendar.day_name[d] for d in range(7)])
```

We see that Friday through Saturay seems to have an increase in logins. If we highlight the weekends, we can try to confirm the case. 

```python
colors = cm.rainbow(np.linspace(0, 1, 7))

pyplot.figure(figsize=(20,10))
pyplot.margins(x=0)
pyplot.xlim(login_agg['login_time'].min(), login_agg['login_time'].max())
for i in range(0,7):
    days = login_agg.loc[login_agg['weekday'] == i]
    if i < 5:
        pyplot.scatter(days.login_time, days['count'], color = colors[0])
    else:
        pyplot.scatter(days.login_time, days['count'], color = colors[1])

pyplot.title("Logins With Weekends Highlighted")
```

For the most part, it seems that weekends in general have a higher login rate than the rest of the week. Though there seems to be an outlier week during 03-16 through 03-20. The login demand seemed to hold steady throughout the weekdays. We could take a closer look at march.

```python
pyplot.figure(figsize=(20,10))
#Plot March
march_df = login_agg.loc[login_agg['month'] == 3]
pyplot.xlim(march_df['login_time'].min(), march_df['login_time'].max())
for i in range(0,7):
    days = march_df.loc[march_df['weekday'] == i]
    if i < 5:
        pyplot.scatter(days.login_time, days['count'], color = colors[0])
    else:
        pyplot.scatter(days.login_time, days['count'], color = colors[1])
```

This closer look seems to confirm the constant login activity in one of the weeks in march. Another likely pattern would be login activity throughout the day.

```python
#Plot an average day
login_agg.groupby('hour')['count'].mean().plot(figsize=(10,5))
```

On an average day, we can see that logins drop off rapidly between 4am and rise up again around 8am. When most people are probably sleeping. There also seems to be a peak around 12pm when customers might need a ride for lunch, then drops slowly afterwards as they stick to work. 

# Part 2 ‐ Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.
However, a toll bridge, with a two ­way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have
proposed an experiment to encourage driver partners to be available in both cities, by
reimbursing all toll costs.
1. What would you choose as the key measure of success of this experiment in
encouraging driver partners to serve both cities, and why would you choose this metric?
2. Describe a practical experiment you would design to compare the effectiveness of the
proposed change in relation to the key measure of success. Please provide details on:
a. how you will implement the experiment
b. what statistical test(s) you will conduct to verify the significance of the
observation
c. how you would interpret the results and provide recommendations to the city
operations team along with any caveats.

## Solution

It appears that the goal of this experiment is to more evenly distribute driver availability in both cities by having them be less exclusive to each city. With this in mind, the metric I would choose would be the percentage of time each driver spends in each city. For example, a driver might spend 75% of their time in Gotham while only spending 25% of their time in Metropolis. If this experiment were successful, we would see the metric shift to a 50/50 time split. If the experiment failed, we would see the metric stay the same, or possibly get even worse, with the driver spending 100% of their time in Gotham.

To compare the effectiveness of the proposed fare change, I will conduct an A/B test and work through the hypothesis testing framework to apply the one tailed Z-test. My assumption is that drivers who set their home location to a particular city, spends most of their time in that city.  I will also assume that the distribution of time spent in each city follows the normal distribution, with the average a high percentage of time in their home city. For example, a driver with a home set in Gotham, will spend 75% of his time there. 

I will first split the drivers by home city to help control for uneven driver distributions between the cities. For example, it may be possible that 90% of all drivers are located in Gotham which would skew the data towards time spent in Gotham, regardless if time spent in each city has improved. The drivers will also be into a control group and a group that will have free fares across the bridge.

I will then construct my null hypotheses for both groups of drivers in each city. As an example, if control group in Gotham spend 75% of their time in Gotham on average, my null hypothesis would be: The mean time a Gotham driver spends in their city is 75%.  The alternative hypothesis, would be that the average amount of time is greater or less than 75%.

The experiment will commence and data will be collected over a set duration. One week might be a good start. Once sufficient data has been collected, we can test for significance. We can select a significance level of 5% on which a probability threshold below this means the null hypothesis can be rejected. We then construct the rejection region, and calculate the Z-value of each group. If the Z-Value is inside the rejection region, then we know the fare change has made a statistically significant impact to the time a driver has spent in their city. We can interpret this by rejecting the null hypothesis, and say that the free fares have reduced the amount of time spent in their home city. The recommendation would be to implement the free fare system. 

One important caveat about this test would be the possibility that the null hypothesis may not be correct. For example, its possible that the mean time spent in each city is much closer to 50% than initially predicted. In this case, the data might indicate that the free bridge fares have a greater effect than we initially predicted.

## Data Cleaning & Analysis

```python
with open('ultimate_data_challenge.json') as f:
    data = json.load(f)

ultimate_df = pd.DataFrame(data)
ultimate_df.info()
```

```python
#Convert dates to datetime
ultimate_df['last_trip_date'] = pd.to_datetime(ultimate_df['last_trip_date'])
ultimate_df['signup_date'] = pd.to_datetime(ultimate_df['signup_date'])
```

There are some missing values to take care of. 
The simplest solution for the numerical values would be to fill it in with the average.
For the string values, since there aren't many values missing, we can drop those rows. 

```python
ultimate_df['avg_rating_by_driver'].fillna(ultimate_df['avg_rating_by_driver'].mean(), inplace=True)
ultimate_df['avg_rating_of_driver'].fillna(ultimate_df['avg_rating_of_driver'].mean(), inplace=True)
ultimate_df.dropna(axis=0, inplace=True)
```

```python
ultimate_df.head()
```

```python
print(ultimate_df.last_trip_date.min())
print(ultimate_df.last_trip_date.max())
```

Create a new feature that = 1 if a user is retained and = 0 if not. We check this by seeing if the last trip date is after June 1st, 2014. 


```python
ultimate_df['retained'] = np.nan
ultimate_df.loc[ultimate_df['last_trip_date'] >= '2014-06-01', 'retained'] = 1
ultimate_df.loc[ultimate_df['last_trip_date'] < '2014-06-01', 'retained'] = 0

num_retained = ultimate_df['retained'].value_counts()[1]
print("Total number of users: " + str(len(ultimate_df)))
print("Number of adopted users: " + str(num_retained))
print("Adoption rate: " + str(num_retained / len(ultimate_df)))
```

We can also get a general age of the account by subtracting the signup_date from the last_trip date. Its possible that this could be a helpful indicator of how likely an account will adopt. 

```python
ultimate_df['last_trip_date'] = pd.to_datetime(ultimate_df['last_trip_date'], unit='s')
ultimate_df['signup_date'] = pd.to_datetime(ultimate_df['signup_date'], unit='s')
ultimate_df['age'] = ultimate_df['last_trip_date'] - ultimate_df['signup_date']
ultimate_df['age'] = ultimate_df['age'].dt.total_seconds()
ultimate_df['age'].fillna(0, inplace=True)
#Then onehot encode categorical features
ultimate_df = pd.get_dummies(ultimate_df, columns = ['city', 'phone', 'ultimate_black_user'])
```

```python
#Fix the datatypes
convert_dict = {'retained': int,
                'age': int,
}
ultimate_df = ultimate_df.astype(convert_dict)
ultimate_df.head()
```

```python
#Since the goal is to predict if a user is active in the 6th month in the system,
#confirm if every user in the dataset has signed up in januaray. 
ultimate_df['signup_date'].value_counts().sort_index()
```

Every account seemed to have been created in januaray, which means the dataset is full of users in their sixth month of the system. 

```python
#Drop dates used to create the account age feature. 
ultimate_df.drop(['last_trip_date', 'signup_date'], axis=1, inplace=True)
#Also drop redundant one hot encoding values. 
ultimate_df.drop(['ultimate_black_user_False', 'phone_Android'], axis=1, inplace=True)
```

## Predictive Model

The predictive model I will use is the basic DecisionTreeClassifier. Not only does the decision tree implictly perform feature selection, but its easier to interpret since we can visualize the decisions the model have made. We could use these insights to imporve the feature engineering for a more advanced black box models like gradient boosting. 

```python
train_df = ultimate_df
X = ultimate_df.drop(['retained'], axis=1)
y = ultimate_df['retained']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
#Picking max_depth of four for improved interpretability
model = DecisionTreeClassifier(max_depth=4)
cv_scores = cross_val_score(model, X_train, y_train, cv=10)
print(cv_scores)
print("Average 5-Fold CV Score: {}".format(np.mean(cv_scores)))
```

```python
#Create a classificaiton report. 
model.fit(X,y)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

An average accuracy of 95% look very promising. The classificaiton report also indicates robust prediction capabilities for both adopted and non adopted users. We can now look at feature importances and visualize the decision tree to see how it's making the predictions. 

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.plot(kind='barh')
pyplot.show()
```

There seems to be a reliance on the feature giving the amount of time between account creation and last login. Its possible that since every user who logged into the application in June would be adopted, that the model has picked up on this.There also seems to be a very small amount of importance given to 'city_King's Landing' and to android users. We can take a look at how the Decision tree was formed. 

```python
dot_data = StringIO()

export_graphviz(model, out_file=dot_data,  
                feature_names=list(X_train.columns),
                class_names = True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

Keep in mind that the age of an account is given in seconds. With that being said, it appears for the most part, if an account is over a certain age, the user will be adopted. The reverse also seems true. One interesting observation is that users in King's Landing seem to create exceptions to this model. IPhone users and ultimate_black_users in kings landing seem to adopt the platform. Overall though, while this model is relatively accurate, it doesn't tell us much about how to increase adoption rates other than to get users to log in more often. It would be interesting to see the model if age was droped. 

```python
ultimate_df.drop(['age'], axis=1, inplace=True)
X = ultimate_df.drop(['retained'], axis=1)
y = ultimate_df['retained']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
#Picking max_depth of four for improved interpretability
model = DecisionTreeClassifier(max_depth=4)
cv_scores = cross_val_score(model, X_train, y_train, cv=10)
print(cv_scores)
print("Average 5-Fold CV Score: {}".format(np.mean(cv_scores)))
```

```python
model.fit(X,y)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

As expected, we see a drop in the accuracy of predicting adopted users. However its still much better than random guessing. We might be able to get some insights from the remaining features on what might most affect adoption rate. 

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.plot(kind='barh')
pyplot.show()
```

While some features are still effectively unused. Its interesting to now see that average driver rating seems to be the next best predictor.

```python
dot_data = StringIO()

export_graphviz(model, out_file=dot_data,  
                feature_names=list(X_train.columns),
                class_names = True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

## Insights

To summarize the insights from the decision tree models:

1. The best predictive model will take into account the time delta between account creation, and last login. This is probably not very helpful information though, since the actionable insight here would simply to get users to log in more often.

2. If the 'age' information is left out, then we see that the user raiting is the main decision node for the model. Strangely  the first decision of having a raiting above 4.95 will land most people into the non adoption side. One possibility would be that one time users tend to get raited by a single five star. The decision tree later puts most users with above a 4.05 raiting into the adoption bucket. Users who get bad raitings tend to not adopt. This may be an intended feature. 

3. There is also an emphasis on Kings Landing. People who use the app in King's landing tend to be more likely to adopt the platform. Being the largest city of the seven kingsdoms means people might be more likely to need a ride sharing service to get around. 

4. Lastly there seems to be some weight given to the phone type. iPhone users tend to adopt more, which could indicate better iPhone support, or even a buggy android verison. 

Ultimate might leverage these insights by looking into why certain users are being rated poorly by drivers, understanding what is going well in King's Landing and understanding the differences between the iPhone and Android versions of the Ultimate app.
